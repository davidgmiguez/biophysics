## Characteristics of complex systems

### A complex system is "not normal”

In the sense that they do not fit normal gaussian distributions. 

<div class="image123">
    <div class="imgContainer">
        <img src="img/normal.jpg" height="400" width="400"/>

</div>

Lets illustrate this concept of normal, versus not-normal system. Let's for instance, play a game where we throw two dices and compute the sum. Repeat the exèrimets an number of times and plot the results in a histogram, 

<div class="image123">
    <div class="imgContainer">
        <img src="img/dices.jpg" height="400" width="400"/>
    </div>

</div>


```julia
number_of_attemps=10000
total = Array{Float64, 2}(undef, 1, number_of_attemps);
```

```julia
for i=1:number_of_attemps
    value1=rand((1,2,3,4,5,6))
    value2=rand((1,2,3,4,5,6))
    total[i]=value1+value2
end
```

```julia
using Plots
```

```julia
p1=histogram(total[:],bins = range(2,12, step = 1));
title!("Normal")
xlabel!("value")
ylabel!("counts");
```

Now, lets repeat the same experiment, but set the rule that if two dices are similar, do it again. Thsi a sort of simple feedback loop, where the output affect the input. 

```julia
for i=1:number_of_attemps
    value1=rand((1,2,3,4,5,6))
    value2=rand((1,2,3,4,5,6))
    if  value1==value2
        value1=rand((1,2,3,4,5,6))
        value2=rand((1,2,3,4,5,6))
    end  
    total[i]=value1+value2
end
```

```julia
p2=histogram(total[:],bins = range(2,12, step = 1));
title!("not Normal")
xlabel!("value")
ylabel!("counts");
```

```julia
plot(p1,p2,layout=(1,2),legend=false)
```

we can see that the two distributions are clearly different, the left one is more similar to a typical gaussian distribution, while the right one, despite having the same average value, has different  tails.
