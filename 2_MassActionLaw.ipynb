## 6. Definition of the Law of Mass Action 

Now that we know how to obtain the stoichiometric matrices and the order of any given reaction, the next step is to write the differential equations that capture the dynamics, i.e., the changes in time of the amount of all the species involved.  To illustrate this, we will use a generic scheme of interaction in vector notation:

$$\begin{align*}
X_1 + X_2 \overset{k}{\longrightarrow} X_3  \tag{58}\\ 
\end{align*}$$
The stoichometrix matrix is: 

```julia
A=([1,1,0])
println("The order of the 1st reaction is $(sum(A))")
println("The Units of k are s^-1 . k^$(1-(sum(A)))")
```

We will write down the differential equations that account for the dynamics of each species. 
$$
\begin{align}
\frac{\mathrm{d} X_1}{\mathrm{d} t}  &=&  \tag{59}\\
\frac{\mathrm{d} X_2}{\mathrm{d} t} &=& \tag{60}\\
\frac{\mathrm{d} X_3}{\mathrm{d} t} &=&  \tag{61}
\end{align}
$$

The __Mass Action Law__ describes how a chemical reaction takes place under ideal conditions. In brief, it is based on the fact that the speed of any reaction is proportional to the concetration of the reactants involved, which has been discussed previously in this course (te proportionalinty constant is the kinetic rate constant). From here, it follows that the rate of change of any reactant is also proportional to the concetrations of the reactants involved. This velocity depends on the affinity of the reaction and the concentrations of each of the reactants (in any scheme of interactions, the ones at the left of the arrow). The affinity of the reactions is defined by the rate constant `k` of the reaction. The `k` is the proportionality constant that correlates the speed fo the reaction with the concentrations of the species involved, which in this case are $X_1$ and $X_2$:

$$
\begin{align}
\frac{\mathrm{d} X_1}{\mathrm{d} t}  &= - k X_1 X_2 \tag{62}\\
\frac{\mathrm{d} X_2}{\mathrm{d} t} &=- k X_1 X_2 \tag{63}\\
\frac{\mathrm{d} X_3}{\mathrm{d} t} &=  k X_1 X_2\tag{64} 
\end{align}
$$

Lets see if the units are correct:

$$
\begin{align}
\frac{[M]}{[s]}  &= - \frac{1}{[s][M]}[M] [M] \tag{65}\\
\frac{[M]}{[s]}  &= - \frac{1}{[s][M]}[M] [M] \tag{66}\\
\frac{[M]}{[s]}  &=  \frac{1}{[s][M]}[M] [M] \tag{67}\\
\end{align}
$$

For the case where more than one molecule of reactant is involved, we will use the reaction of formation of water. 

```julia
A=[2 1 0]
B=[0 0 2]
stoichiometric_matrix= (B-A)'
println("The order of the reaction is $(sum(A))")
println("The units of k are s^-1 . k^$(1-(sum(A)))")
println("The stoichiometric matrix is $stoichiometric_matrix ")
```

we can decompose the equation as
 $$  X_1 +  X_1 + X_2 \overset{k}{\longrightarrow}  X_3 + X_3 \tag{68}$$
We will write down the differential equations that account for the dynamics of each species. The speed of each reaction depends on the concentration of the species involved (what is before the arrow). In this case is two units of $X_1$ and one unit of $X_2$. In addition, since the reaction consumes 2 units of $X_1$ for each unit of $X_2$, the rate of consumption of $X_1$ is twice the rate of consumption $X_2$. Another more automaticic way of doing it is to use teh matrix of stochoimetric coefficients $(B-A)'$. Therefore:
$$
\begin{align}
\frac{\mathrm{d} X_1}{\mathrm{d} t}  &= - 2k X_1^2 X_2 \tag{69}\\
\frac{\mathrm{d} X_2}{\mathrm{d} t} &=- k X_1^2 X_2 \tag{70}\\
\frac{\mathrm{d} X_3}{\mathrm{d} t} &=2  k X_1^2 X_2 \tag{71}
\end{align}
$$
Lets check if units are correct:
$$
\begin{align}
\frac{[M]}{[s]}  &=& - \frac{1}{[s][M]^2} [M]^2 [M] \tag{72}\\
\frac{[M]}{[s]}  &=& - \frac{1}{[s][M]^2} [M]^2 [M] \tag{73}\\
\frac{[M]}{[s]}  &=& \frac{1}{[s][M]^2} [M]^2 [M] \tag{74}
\end{align}
$$


## 6. General formulation of the Mass Action Law

When we have to write the differenetial equations of more complex systems of interactions, such as Eq 21, we can use a general method based omn the following equation: 
$$
\begin{align}
\frac{\mathrm{d} X}{\mathrm{d} t}&= (B-A)^T \cdot K \cdot X^A \tag{75}\\
\end{align}
$$
where  `A` and `B` are the matrices with the stoichiometric coefficients, `X` is the state vector $X=[X_1,X_2,X_3]^T$, and `K` is a matrix in the form:

$$
K=\begin{pmatrix}
 k_1 & 0   & 0  & \dots & 0 \\ 
 0 &  k_2  & 0  & \dots & 0\\ 
 0 & 0  & k_3 &\dots  &0 \\ 
 \vdots &  \vdots   &  \vdots  &  \vdots  &  \vdots  \tag{76}\\ 
 0 & 0  & 0  &  \dots & k_r 
\end{pmatrix}
$$

$X^A$ is a column vector, calculated as:

$$X^A=\begin{pmatrix}
X_1^{A_{11}} \cdot X_2^{A_{12}} \cdot X_3^{A_{13}} \cdot  ... \cdot  X_s^{A_{1s}} \\ 
X_1^{A_{21}} \cdot X_2^{A_{22}} \cdot X_3^{A_{23}} \cdot  ... \cdot  X_s^{A_{2s}} \\ 
 \vdots  \,\,\,\,\,\,\,\,\,\,\,  \vdots  \,\,\,\,\,\,\,\,\,\,\,\,    \vdots \,\,\,\,\,\,\,\,\,   \vdots  \,\,\,\,\,\,\,\,   \vdots  \\ 
X_1^{A_{r1}} \cdot X_2^{A_{r2}} \cdot X_3^{A_{r3}} \cdot  ... \cdot  X_s^{A_{rs}} \\ 
\end{pmatrix} =\begin{pmatrix}
 \prod_{i=1}^{s} X_{i}^{A_{1i}}\\ 
\prod_{i=1}^{s} X_{i}^{A_{2i}}\\ 
   \vdots  \tag{77} \\ 
\prod_{i=1}^{s} X_{i}^{A_{ri}}\\ 
\end{pmatrix}
$$


#### Example of differential form of Mass Action Law

As an example we will use the previous reaction for the formation of salt. 

$$NaCO_3 + CaCl_2 \overset{k_1}{\underset{k_2}{\longleftrightarrow}} CaCO_3 + 2 \cdot NaCl \tag{78}\\  $$

with two reactions $r=2$ and four species $s=4$, with the stoichiometric matrices `A` and `B` listed in Eq. 30

```julia
A=[1 1 0 0; 0 0 1 2]
B=[0 0 1 2; 1 1 0 0]
stoichiometric_matrix= (B-A)'
println("The stoichiometric matrix is $stoichiometric_matrix ")
```

in our case

$$
K=\begin{pmatrix}
 k_1 & 0   \tag{79}\\ 
 0 &  k_2  \\ 
\end{pmatrix}
$$

and 

$$X^A=\begin{pmatrix}
X_1\cdot X_2  \tag{80}\\ 
X_3\cdot X_4^2  
\end{pmatrix}
$$

therefore

$$
\begin{align}
 \begin{bmatrix}
\frac{\mathrm{d} X_1}{\mathrm{d} t}\\ \frac{\mathrm{d} X_2}{\mathrm{d} t} \\ \frac{\mathrm{d} X_3}{\mathrm{d} t} \\ \frac{\mathrm{d} X_4}{\mathrm{d} t}
\end{bmatrix}&= \begin{bmatrix}
 -1   &  1 \\
    -1  &   1\\
     1  &  -1\\
     2  &  -2
\end{bmatrix}\begin{bmatrix}
k_1 & 0\\ 
 0& k_2
\end{bmatrix}\begin{bmatrix}
X_1 \cdot X_2\\ 
X_3 \cdot X_4^2
\end{bmatrix} \tag{81}
\end{align}
$$

so calculating

$$
\begin{align}
\begin{bmatrix}
\frac{\mathrm{d} X_1}{\mathrm{d} t}\\ \frac{\mathrm{d} X_2}{\mathrm{d} t} \\ \frac{\mathrm{d} X_3}{\mathrm{d} t} \\ \frac{\mathrm{d} X_4}{\mathrm{d} t}
\end{bmatrix}&= \begin{bmatrix}
 -1   &  1 \\
    -1  &   1\\
     1  &  -1\\
     2  &  -2
\end{bmatrix}\begin{bmatrix}
k_1 \cdot  X_1 \cdot X_2\\ 
k_2 \cdot X_3 \cdot X_4^2
\end{bmatrix} \tag{82}
\end{align}
$$

In other words, the differential equations are obtained by multiplying the matrix $(B-A)'$ by a vector of speeds of the reactions involved. This way, the final set of equations is:

$$\begin{align}        
            \frac{ \mathrm{d}X_1 }{\mathrm{d}t} &= - k_1 \cdot X_1 \cdot X_2 + k_2 \cdot X_3 \cdot X_4^2  \tag{83}\\ 
            \frac{ \mathrm{d}X_2 }{\mathrm{d}t} &= - k_1 \cdot X_1 \cdot X_2 + k_2 \cdot X_3 \cdot X_4^2   \tag{84}  \\
             \frac{ \mathrm{d} X_3 }{\mathrm{d}t} &=  k_1 \cdot X_1 \cdot X_2 - k_2 \cdot X_3 \cdot X_4^2  \tag{85}  \\
              \frac{ \mathrm{d} X_4 }{\mathrm{d}t} &= 2 k_1 \cdot X_1  \cdot X_2 - 2 k_2 \cdot X_3 \cdot X_4^2   \tag{86}  \\
\end{align}
$$

## 7. Mass Conservation Law

The __Law of Conservation of Mass__ or __principle of Mass Conservation__ states that for any system closed to all transfers of matter and energy, the mass of the system must remain constant over time, as the system's mass cannot change, so quantity can neither be added nor be removed. Therefore, the quantity of mass is conserved over time. For example, in chemical reactions, the mass of the chemical components before the reaction is equal to the mass of the components after the reaction. Thus, during any chemical reaction and low-energy thermodynamic processes in an isolated system, the total mass of the reactants, or starting materials, must be equal to the mass of the products. 

We can use this law to find relationships between species of a given reaction, in such a way that we can reduce the number of variables of the systems. The mathematical formulation of this law is the following:

Suppose we have a solution of the system of equations defined by the stoichiometric matrix in the form $C \cdot (B-A)^T =0$ such that $C$ is a vector with number of rows $j= 1...s$ equal to the number of species involved `s`, then the following is true:

$$
\begin{align}
 C_1  X_1 + C_2 X_2 + ... + C_s  X_s = cte \tag{87}
\end{align}
$$

this equation is a mass conservation law of the reaction. 

$$
\begin{align}
 C_1 X_1 + C_2 X_2 + ... + C_s  X_s = cte \tag{88}
\end{align}
$$

this is true because, if we take the derivative of the last equation:
$$
\begin{align}
\frac{\mathrm{d} [C_1 X_1 + C_2  X_2 + ... + C_s  X_s ]}{\mathrm{d} t}&= \frac{\mathrm{d} [C_1]  X_1}{\mathrm{d} t} + \frac{\mathrm{d} [C_2]  X_2}{\mathrm{d} t} +...+ \frac{\mathrm{d} [C_s]  X_s}{\mathrm{d} t}  \tag{89}
\end{align}
$$
since $C$ is a vector of constant values we can write:
$$
\begin{align}
[C_1] \frac{\mathrm{d}   X_1}{\mathrm{d} t} + [C_2] \frac{\mathrm{d} X_2}{\mathrm{d} t} +...+ [C_s] \frac{\mathrm{d}  X_s}{\mathrm{d} t} = \begin{bmatrix}
C_1 & C_2 & ... & C_s \end{bmatrix} \begin{bmatrix}
\frac{\mathrm{d} X_1}{\mathrm{d} t}\\ \frac{\mathrm{d} X_2}{\mathrm{d} t} \\ \frac{\mathrm{d} X_3}{\mathrm{d} t} \\ \vdots  \\ \frac{\mathrm{d} X_s}{\mathrm{d} t} \tag{90}
\end{bmatrix}
\end{align}
$$
Now we know that, based on the general equation of the differentiatl form of mass action 

$$
\begin{align}
 \begin{bmatrix}
C_1 & C_2 & ... & C_s \end{bmatrix} \begin{bmatrix}
\frac{\mathrm{d} X_1}{\mathrm{d} t}\\ \frac{\mathrm{d} X_2}{\mathrm{d} t} \\ \frac{\mathrm{d} X_3}{\mathrm{d} t} \\ \vdots  \\ \frac{\mathrm{d} X_s}{\mathrm{d} t} \end{bmatrix} =  \begin{bmatrix}
C_1 & C_2 & ... & C_s \end{bmatrix}  (B-A)^T \cdot K \cdot X^A \tag{91}  
\end{align}
$$ 

we also know that  $C \cdot (B-A)^T =0$ by definition (vector C was a solution of the system of equations), so

$$
\begin{align}
 \begin{bmatrix}
C_1 & C_2 & ... & C_s \end{bmatrix} \begin{bmatrix}
\frac{\mathrm{d} X_1}{\mathrm{d} t}\\ \frac{\mathrm{d} X_2}{\mathrm{d} t} \\ \frac{\mathrm{d} X_3}{\mathrm{d} t} \\ \vdots  \\ \frac{\mathrm{d} X_s}{\mathrm{d} t} \end{bmatrix} =  0 \cdot K \cdot X^A \tag{92}  = 0 
\end{align}
$$ 

therefore 
$$
\begin{align}
 C_1 \cdot X_1 + C_2 \cdot X_2 + ... + C_s \cdot X_s = cte \tag{93}
\end{align}
$$


### Example of Mass Conservation Law

Lets now apply the theory to find conservation laws in the reaction of the example.

$$NaCO_3 + CaCl_2 \overset{k_1}{\underset{k_2}{\longleftrightarrow}} CaCO_3 + 2 \cdot NaCl \tag{94}\\  $$

we have to find  $C \cdot (B-A)^T =0$ with $C$ being a vector with $j= 1...s$, being $s=4$. 

$$
\begin{bmatrix}
 C_1   &  C_2  &  C_3  &  C_4 \end{bmatrix}\begin{bmatrix}
 -1   &  1 \\
    -1  &   1\\
     1  &  -1\\
     2  &  -2
\end{bmatrix}=0 \tag{95}
$$

So we have to solve this system of equations:

$$\begin{cases} -C_1 - C_2 + C_3 + 2 \cdot C_4 = 0 \\ C_1 + C_2 - C_3 - 2 \cdot C_4 = 0 \end{cases} \tag{96}$$

since the two reactions correspond to a reversible system, the second equation is simply the first one multiplied by $-1$. Therefore the solytion of the system is simply
$$
C_3 + 2 \cdot C_4 - C_2 = C_1 \tag{97}
$$

so, we obtain one equation and three unknown variables. If we give values for three of the constants ($C_{1 \dots 3}=1$) and calculate the other one, we obtain:

$$
\begin{align}
1 + 2 \cdot C_4 - 1 &= 1 \tag{98}\\
C_4&=\frac{1}{2}\tag{94}
\end{align}
$$

the following the equation for the mass conservation, we obtain:


$$
\begin{align}
 X_1 + X_2 + X_3 +\frac{X_4}{2}= cte \tag{99}
\end{align}
$$

Let's test if this is correct by solving numerically the system (we will learn later how to do this). 


```julia
using DifferentialEquations
using ParameterizedFunctions
using Plots 
```

```julia
simpleODE3! = @ode_def abetterway begin
  da = -k1 * a * b + k2 * c * d^2
  db = -k1 * a * b + k2 * c * d^2
  dc = k1 * a * b - k2 * c * d^2
  dd = 2 * k1 * a * b - 2 * k2 * c * d^2
    end k1 k2
```

```julia
k1=2.3e0;  # units 1/(Ms)
k2=2.5e0;  # units 1/(M M s)

p = (k1,k2)

a₀=0.02; # units (M)
b₀=0.01; # units (M)
c₀=0; # units (M)
d₀=0; # units (M)

u₀=[a₀,b₀,c₀,d₀];
```

```julia
tspan = (0.0,300.0)
prob = ODEProblem(simpleODE3!,u₀,tspan,p)
sol = solve(prob)
plot(sol,title= "Solution of differential equations",xlabel="Time [s]",ylabel="Concentration [M]")
plot!(sol.t,sol[1,:].+sol[2,:].+sol[3,:].+0.5.*sol[4,:],label="conservation")
```

## 9. Conclusions

The __Law of Mass Action__ or __Mass Action Law__ was initially derived in the context of chemical reactions, it can be used also to describe the equilibirum of any system of interacting entities (rabbits and foxes, bacterias and T cells, genes and proteins, humans...).

In brief, it allows us to calculate the dynamics of a set of interacting entitites of a system, and ic constitutes the most widely used tool to study biological systems from a quantitative predictive perspective. When possible, the __Mass Conservation Law__ allows us to reduce the complexity of the system by reducing the number of independent variables of the system of equations derived from the Mass Action Law, by providing us with dependencies between the variables of the system, allowing us to simplify the system and reduce the number of equations to solve. This search for mass conservation laws becomes more useful once whe have systems of more that one reversible reactions.
