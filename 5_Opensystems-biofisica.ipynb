# Open Systems and Nonlinear Interactions


## 1. Mass Action Law for Open Systems

Despite being developed in the context of closed chemical systems, the __Mass Action Law__ is also applicable to open systems, i.e., systems where there is no consevarion of mass. This is a special case in chemistry (open chemical reactors), but extremely relevanty for biological systems, since biological species are are often part of  open systems that exchanges mass, energy or information with the environment. Let's explore how the Mass action takes into account these type of open systems. The first and most simple example of open system is when a given reactant is degraded or flows away. This is illustrated as the following interaction scheme:

$$   X_1  \overset{k_1}{\longrightarrow} 0 \tag{1} $$ 

with initial concentration $X_1(0)$ and stoichometric matrices

$$
A=\begin{bmatrix}
 1\end{bmatrix} ;
B=\begin{bmatrix}
0   
\end{bmatrix} ; (B-A)^T= \begin{bmatrix}
 - 1 \end{bmatrix} \tag{2}
$$
 
in this particular case

$$
K=\begin{pmatrix}
 k_1\\ 
\end{pmatrix} \tag{3}
$$

and 

$$X^A=\begin{pmatrix}
X_1^1
\end{pmatrix} \tag{4}
$$

so 

$$
\begin{align}
 \begin{bmatrix}
\frac{\mathrm{d} X_1}{\mathrm{d} t} \end{bmatrix}= \begin{bmatrix}
 - 1 \end{bmatrix} \begin{bmatrix}
k_1 \end{bmatrix}\begin{pmatrix}
X_1
\end{pmatrix} = \begin{bmatrix}
 - 1 \end{bmatrix} \begin{pmatrix}
k_1  X_1
\end{pmatrix} = \begin{bmatrix}
 - k_1 X_1 \end{bmatrix}  \tag{5}
\end{align}
$$

So the ODE is simply.

$$\frac{\mathrm{d} X_1}{\mathrm{d} t} =  - k_1 X_1 \tag{6}$$ 

with a solution that is equivalent to the irreversible first order reaction.
 
 $$ X_1 (t)= X_1 (0) e^{- k_1 t}  \tag{7}$$
 

```julia
using DifferentialEquations
using ParameterizedFunctions
using Plots 
gr()
```

```julia
X₀ = [0.1,0.0]
t₀=0
t_final=10
t=collect(t₀:0.1:t_final) # time
k_1 = 1.
plot(t,t->X₀[1]*exp(-k_1*t),label="\\ X_1")
title!("Degradation first order")
xaxis!("Time")
yaxis!("Concentration")
```

Another simple example is the generation of a given reactant in the system, such as:
$$   0  \overset{k_1}{\longrightarrow} X_1 \tag{8} $$ 

with initial concentration $X_1(0)$ and stoichometric matrices

$$
A=\begin{bmatrix}
 0\end{bmatrix} ;
B=\begin{bmatrix}
1   
\end{bmatrix} ; (B-A)^T= \begin{bmatrix}
 1 \end{bmatrix} \tag{9}
$$
 
in this particular case

$$
K=\begin{pmatrix}
 k_1\\ 
\end{pmatrix} \tag{10}
$$

and 

$$X^A=\begin{pmatrix}
X_1^0
\end{pmatrix}=\begin{pmatrix}
1
\end{pmatrix} \tag{11}
$$

so 

$$
\begin{align}
 \begin{bmatrix}
\frac{\mathrm{d} X_1}{\mathrm{d} t} \end{bmatrix}= \begin{bmatrix}
 1 \end{bmatrix} \begin{bmatrix}
k_1 \end{bmatrix}\begin{pmatrix}
1
\end{pmatrix}  \tag{12}
\end{align}
$$

So, sor this particular case, the corresponding differential equation is simply:

 $$\frac{\mathrm{d} X_1}{\mathrm{d} t} =  k_1 \tag{13}$$ 
 
 which when we integrate we obtain the following solution:
 
$$\begin{align}
\int \mathrm{d} X_1 &= \int k_1  \mathrm{d} t  \tag{14}\\
 X_1(t) &=  k_1  \cdot t + C \tag{15}
\end{align} $$

again, to calculate the constant $C$ we use the initial condition for $t=0$, so $ X_1(0)=C$. Therefore:
$$ X_1(t) =  k_1  \cdot t + X_1(0) \tag{16}$$ 
which shows that the amount of $X_1$ grows linearly with $t$.

```julia
plot(t,t->X₀[1]+k_1*t,label="\\ X_1")
title!("Production zero order")
xaxis!("Time")
yaxis!("Concentration")
```

Let's now explore some special cases of open systems that look in principle very similar to the previous one:

$$   X_1  \overset{k_1}{\longrightarrow} 2 X_1 \tag{17} $$ 

and this one

$$   2X_1  \overset{k_1}{\longrightarrow} 3 X_1 \tag{18} $$ 

These two reaction schemes seem very similar to the previous case of linear producton, since for every reaction you end up with a new molecule of your reactant $X_1$, so we expect a similar dynamics. On the contrary, when we write the differential equations for these two cases we have:


$$
A=\begin{bmatrix}
 1\end{bmatrix} ;
B=\begin{bmatrix}
2   
\end{bmatrix} ; (B-A)^T= \begin{bmatrix}
 1 \end{bmatrix} \tag{19}
$$
 
and for the second order production case
$$
A=\begin{bmatrix}
 2\end{bmatrix} ;
B=\begin{bmatrix}
3   
\end{bmatrix} ; (B-A)^T= \begin{bmatrix}
 1 \end{bmatrix} \tag{20}
$$

In fact, $(B-A)^T$ has the same value in the three cases, but $X^A$ does not. For the first order production

$$X^A=\begin{pmatrix}
X_1^1
\end{pmatrix} \tag{21}
$$

and for the second order

$$X^A=\begin{pmatrix}
X_1^2
\end{pmatrix} \tag{22}
$$

so, following the Mass acton general form, for the first order productionreaction:

 $$\frac{\mathrm{d} X_1}{\mathrm{d} t} =  k_1 \cdot X_1 \tag{23}$$  
 
 and for the second order production 
 
 $$\frac{\mathrm{d} X_1}{\mathrm{d} t} = k_1 \cdot X_1^2 \tag{24}$$ 
 
 
 We know the solution for the first order production:
 
  $$ X_1 (t)= X_1 (0) \cdot e^{k_1 \cdot t}  \tag{25}$$
 
withe the second one is :

$$\begin{align}
\int \frac{\mathrm{d} X_1}{X_1^2} &=  \int k_1 \cdot \mathrm{d}t \tag{26} \\
-X_1^{-1} &=  k_1 \cdot t + C \tag{27}\\
X_1(t) &=\frac{-1}{k_1 \cdot t + C}  \tag{28}
\end{align}$$

using the intial condition of $X_1(t=0)$ we obtain the value of the integration constant. 
$$\begin{align}
X_1(0) &=\frac{-1}{k_1 \cdot 0 + C} = \frac{-1}{C}  \tag{29}\\
C &= \frac{-1}{X_1(0)}  \tag{30}\\
\end{align}$$

so the final equation of is
$$\begin{align}
X_1(t) &=\frac{-1}{k_1  t - \frac{1}{X_1(0)} } \tag{31}
\end{align}$$
so, multiplying numerator and denominator of the right side of the equation by $-X_1(0)$ we obtain
$$\begin{align}
X_1(t) &=\frac{(-1)\cdot(-X_1(0))}{(-X_1(0))(k_1 \cdot t  - \frac{1}{X_1(0)}) }= \frac{X_1(0)}{(1-X_1(0)k_1\cdot t)}\tag{32}
\end{align}$$

```julia
X₀ = [0.5,0.0]
k_1 = 0.1
plot(t,t->X₀[1]+k_1*t,label="\\ zero order reaction \\ X_1 ")
k_1 = 0.1
plot!(t,t->X₀[1]*exp(k_1*t),label=" first order reaction \\ X_1 ")
k_1 = 0.1
plot!(t,t->X₀[1]/(1-k_1*t*X₀[1]),label=" Second order reaction \\ X_1 ")
title!("Production  first order")
xaxis!("Time")
yaxis!("Concentration")
```

We can see that, despite being similar in terms of the production of molecules of $X_1$ per reaction, the dynamics of zero, first and second order production of a given species are completely different, even when using identical parameter values. Also, the amount of $X-1$ produced for the first and second order reactions depends strongly on the initial condition.

```julia
X₀ = [0.75,0.0]
k_1 = 0.1
plot(t,t->X₀[1]+k_1*t,label="\\ zero order reaction \\ X_1 ")
k_1 = 0.1
plot!(t,t->X₀[1]*exp(k_1*t),label=" first order reaction \\ X_1 ")
k_1 = 0.1
plot!(t,t->X₀[1]/(1-k_1*t*X₀[1]),label=" Second order reaction \\ X_1 ")
title!("Production  first order")
xaxis!("Time")
yaxis!("Concentration")
```

This first and second order production are basic examples of a nonlinear system. In this case, these type of system are defined as autocatalytic, where the production of a given species depends on the amount of the same species. This again is quite difficult to picture in the context of a chemical reaction. On the other hand, nature is full of systems where autocalysis is a main feature or, systems that are open in the sense that they can exchange mass, energy or information with their surroundings (for instance, a cell). From the Physics point of view, this constant supply or removal of a species in a system of interacting entities can push it away from its thermodinamic equilibrium, driving very important features that are essential for live itself (as we will see later in the course).  

## Example of Open System: The Brusselator

The Brusselator is a simplified model for a very famous chemical reaction, the __Belousov-Zhabotinsky__ reaction. This reactions has presented a very impornat challenge in the fields of Chemistry and Physics, because it was thought that it broke the second principle of Thermodynamics. You can find more History about the BZ system here. The reaction scheme is as follows:
$$
\begin{align}
0  &\overset{k_1}{\underset{k_2}{\longleftrightarrow}} x \tag{33} \\
x &\overset{k_3}\longrightarrow y \tag{34}\\
2x + y &\overset{k_4}\longrightarrow 3x \tag{35}\\
\end{align}
 $$

where `x, b, y, d` are the reactants species. We will use the general formulation of the mass action in differential form to derive the differentiatl equations that govern the dynamics of the system. The state vector can be defined as $X= [x,b,y,d]$.
$$
\begin{align}
\frac{\mathrm{d} X}{\mathrm{d} t}&= (B-A)^T \cdot K \cdot X^A \tag{36}\\
\end{align}
$$
 
First, we write the matrices of stoichiometric coefficients:

$$
A=\begin{bmatrix}
  0 & 0  \\  
  1 & 0   \\
  1 & 0   \\
  2 & 1    \end{bmatrix} ;
B=\begin{bmatrix}
  1 & 0   \\ 
  0 & 0   \\
  0 & 1   \\
  3 & 0     \end{bmatrix}; \tag{37}
$$

```julia
A = [0 0;1 0;1 0 ;2 1]
B = [1 0;0 0;0 1;3 0]
stoichiometric_matrix= (B-A)'
println("The stoichiometric matrix is $stoichiometric_matrix ")
```

in this particular case

$$
K=\begin{pmatrix}
 k_1 & 0 & 0 & 0  \tag{38}\\ 
 0 &  k_2 & 0 & 0 \\ 
 0 &  0 & k_3 & 0 \\ 
 0 &  0 & 0 &  k_4 
\end{pmatrix}
$$

and 


$$X^A=\begin{pmatrix}
X_1^0\cdot X_2^0  \\
X_1^1\cdot X_2^0  \\
X_1^1\cdot X_2^0  \\
X_1^2\cdot X_2^1  
\end{pmatrix} = \begin{pmatrix}
1 \\
 X_1 \\
 X_1 \\
 X_1^2 \cdot X_2
\end{pmatrix} \tag{39}
$$

so, the equations that define the system are

$$
\begin{align}
 \begin{bmatrix}
\frac{\mathrm{d} X_1}{\mathrm{d} t}\\ \frac{\mathrm{d} X_2}{\mathrm{d} t} \end{bmatrix}& 
=  \begin{bmatrix} 1 & -1 & - 1 & 1  \\ 0 & 0 &  1 & -1  \end{bmatrix}
\begin{pmatrix}
 k_1 & 0 & 0 & 0  \\ 
 0 &  k_2 & 0 & 0 \\ 
 0 &  0 & k_3 & 0 \\ 
 0 &  0 & 0 &  k_4 
\end{pmatrix}
\begin{pmatrix}
1 \\
 X_1 \\
 X_1 \\
 X_1^2 \cdot X_2
\end{pmatrix} 
\end{align} \tag{40}
$$
and after multiplying the matrices
$$
\begin{align}
 \begin{bmatrix}
\frac{\mathrm{d} X_1}{\mathrm{d} t}\\ \frac{\mathrm{d} X_2}{\mathrm{d} t} \end{bmatrix}& 
=  \begin{bmatrix} 1 & -1 & - 1 & 1  \\ 0 & 0 &  1 & -1  \end{bmatrix}
\begin{pmatrix}
k_1 \\
 k_2 \cdot X_1 \\
 k_3 \cdot X_1 \\
 k_4 \cdot X_1^2 \cdot X_2
\end{pmatrix}  \tag{41}
\end{align}
$$


so finally, 

$$
\begin{align}
 \begin{bmatrix}
\frac{\mathrm{d} X_1}{\mathrm{d} t}\\ \frac{\mathrm{d} X_2}{\mathrm{d} t} 
\end{bmatrix}&= 
 \begin{pmatrix}
 k_1  - k_2 \cdot X_1 - k_3 \cdot X_1 +  k_4 \cdot X_1^2 \cdot X_2  \\
 k_3 \cdot X_1 -  k_4 \cdot X_1^2 \cdot X_2 
\end{pmatrix} \tag{42}
\end{align}
$$

Therefore, the  equations for the evolution of `[x]` and `[y]` are as follows:

$$\begin{align}       
            \frac{ d[x] }{dt} &= k_1  - (k_2+k_3) \cdot x + k_4 \cdot x^2 \cdot y = f(x,y)\tag{43} \\ 
            \frac{ d[y] }{dt} &= k_3 \cdot x -  k_4 \cdot x^2 \cdot y =g(x,y) \tag{44}   
            \end{align}            $$

We assume as initial conditions:

$$\begin{align}       
            [x (0)] &= 0 \tag{45} \\ 
            [y (0)] &= 0   \tag{46}   
            \end{align}            
            $$
           
To calculate the equilibrium, we just set the two equations to zero and solve for `X` and `Y`. The first solution from the second equation is $[[x]_{eq}]=0$. The other solution is
$$\begin{align}  
[x]_{eq}(k_3-k_4[x]_{eq}[y]_{eq})&=0 ) \tag{47}  \\
k_3-k_4[x]_{eq}[y]_{eq}&=0  \tag{48}  \\
[x]_{eq}&=\frac{k_3}{k_4}\frac{1}{[y]_{eq}} \tag{49}  
\end{align}
$$

subtsituing this into the first equation of the brusselator, 
$$\begin{align} 
k_1  - (k_2+k_3) \cdot \frac{k_3}{k_4}\frac{1}{[y]_{eq}} + k_4 \cdot \frac{k_3^2}{k_4^2}\frac{1}{[y]_{eq}^2} \cdot [y]_{eq}&=0  \tag{50}  \\
k_1  - (k_2+k_3) \cdot \frac{k_3}{k_4}\frac{1}{[y]_{eq}} + \frac{k_3^2}{k_4}\frac{1}{[y]_{eq}} &=0  \tag{51}  \\
(k_2+k_3) \cdot \frac{k_3}{k_4}\frac{1}{[y]_{eq}} - \frac{k_3^2}{k_4}\frac{1}{[y]_{eq}} &=k1 \tag{52}  \\
(k_2+k_3) \cdot \frac{k_3}{k_4}- \frac{k_3^2}{k_4} &=k1 \cdot [y]_{eq} \tag{53}  \\
\frac{k_2 \cdot k_3}{k_4}+\frac{k_3^2}{k_4}) - \frac{k_3^2}{k_4} &=k1 \cdot [y]_{eq} \tag{54}  \\
\frac{k_2 \cdot k_3}{k_4} &=k1 \cdot [y]_{eq} \tag{55}  \\
\frac{k_2 \cdot k_3}{k_4 \cdot k1} &= [y]_{eq} \tag{56}  
\end{align}            $$

so, subtituing into the equation of $[x]_{eq}$, we obtain the equilibrium point 
$$\begin{align} 
            [x]_{eq} &= \frac{k_1}{k_2}\tag{57} \\ 
            [y]_{eq} &= \frac{k_3  k_2}{k_4  k_1} \tag{58}   
            \end{align}            $$
            
To evaluate stability, we can evaluate the Jacobian at the stationary state $([x]_{eq},[y]_{eq})$. The equilibrium point is an attractor if teh eigenvalues Re($\lambda$)<0, and is a reppellor if Re($\lambda$)>0,.i.e, if perturbations decay or get amplified.

$$
\begin{align}
 J=\begin{bmatrix} 
 \frac{\mathrm{d}f(x,y) }{dx} & \frac{\mathrm{d}f(x,y)}{dy} \\ 
 \frac{ \mathrm{d}g(x,y) }{dx} & \frac{ \mathrm{d}g(x,y) }{dy}
 \end{bmatrix}_{eq} = \begin{bmatrix} 
 2 k_4 \cdot x \cdot y -(k_2+k_3)  &  k_4 \cdot x^2    \\ 
k_3  -  2 k_4 \cdot x \cdot y   & -  k_4 \cdot x^2 
 \end{bmatrix} \tag{59}  
 \end{align} 
$$

substituting the values of the equilibriun for $x$ and $y$


$$
\begin{align}
 J=\begin{bmatrix} 
 \frac{\mathrm{d}f(x,y) }{dx} & \frac{\mathrm{d}f(x,y)}{dy} \\ 
 \frac{ \mathrm{d}g(x,y) }{dx} & \frac{ \mathrm{d}g(x,y) }{dy}
 \end{bmatrix}_{eq} = \begin{bmatrix} 
 2 k_4 \cdot \frac{k_1}{k_2} \cdot \frac{k_3  k_2}{k_4  k_1} -(k_2+k_3)  &  k_4 \cdot \frac{k_1^2}{k_2^2}    \\ 
k_3  -  2 k_4 \cdot \frac{k_1}{k_2} \cdot \frac{k_3  k_2}{k_4  k_1}   & -  k_4 \cdot \frac{k_1^2}{k_2^2}  
 \end{bmatrix} \tag{60}  
 \end{align} 
$$

and simplifying 

$$
\begin{align}
 J=\begin{bmatrix} 
 \frac{\mathrm{d}f(x,y) }{dx} & \frac{\mathrm{d}f(x,y)}{dy} \\ 
 \frac{ \mathrm{d}g(x,y) }{dx} & \frac{ \mathrm{d}g(x,y) }{dy}
 \end{bmatrix}_{eq} = \begin{bmatrix} 
 k_3 -(k_2+k_3)  &  k_4 \cdot \frac{k_1^2}{k_2^2}      \\ 
k_3  -  2 k_3  & -  k_4 \cdot \frac{k_1^2}{k_2^2}   
 \end{bmatrix}=\begin{bmatrix} 
 -k_2   &  k_4 \cdot \frac{k_1^2}{k_2^2}     \\ 
 - k_3  & -  k_4 \cdot \frac{k_1^2}{k_2^2}   
 \end{bmatrix} \tag{61}  
 \end{align} 
$$

To find the characteristic polynomial we do $J-\lambda I$

$$
\begin{align}
 J-\lambda I=\begin{bmatrix} 
 \frac{\mathrm{d}f(x,y) }{dx}-\lambda & \frac{\mathrm{d}f(x,y)}{dy} \\ 
 \frac{ \mathrm{d}g(x,y) }{dx} & \frac{ \mathrm{d}g(x,y) }{dy}-\lambda
 \end{bmatrix}_{eq} =\begin{bmatrix} 
 -k_2    - \lambda &  k_4 \cdot \frac{k_1^2}{k_2^2}     \\ 
 - k_3  & -  k_4 \cdot \frac{k_1^2}{k_2^2}   - \lambda 
 \end{bmatrix} \tag{62}  
 \end{align} 
$$

to calculate the eigenvalues $det[J-\lambda I]=0$

$$
\begin{align}
( -k_2    - \lambda)(-  k_4 \cdot \frac{k_1^2}{k_2^2}   - \lambda )-( - k_3 \cdot k_4 \cdot \frac{k_1^2}{k_2^2}  ) &=0\tag{63}  \\
\lambda^2+\lambda(k_2+k_4\frac{k_1^2}{k_2^2})+k_4\frac{k_1^2}{k_2^2}(k_3+k_2) &=0 \tag{64}  
\end{align} $$

Lets now use some values for the parameters and test numerically the system:

```julia
k_1=1.
k_2=1.
k_3=1.
k_4=1.;
```

The value at the steady state is.

```julia
 [k_1/k_2, k_3*k_2/k_4/k_1]
```

Lets plot now the nullclines. 

```julia
x₀=0.5
x_final=4
x_=collect(x₀:0.1:x_final) # time
p1=plot(x_,x_->k_3/k_4/x_,label="nullcline 1")
P1=plot!(x_,x_-> (k_2+k_3) / k_4 /x_ - k_1/k_4/x_ / x_ ,label="nullcline 2 ")

title!("nullclines")
xaxis!("Time")
yaxis!("Concentration")
```

Now lets look at the stability of the steady state, by looking a the eigenvalues

```julia
a=1.0
b=k_2+k_4*(k_1^2/k_2^2)
c=(k_1^2/k_2^2)*k_4*(k_3+k_2)
```

```julia
function quadratic(a, b, c)
          discr = b^2 - 4*a*c
          discr >= 0 ?   ( (-b + sqrt(discr))/(2a), (-b - sqrt(discr))/(2a) ) : error("Only complex roots")
        end
```

```julia
quadratic(a,b,c)
```

The two values are imaginary. 

Finally, we solve numerically the system 

```julia
using DifferentialEquations
using ParameterizedFunctions
using Plots;  gr()
```

```julia
brusselator3! = @ode_def BR begin
    dX = k_1  - (k_2+k_3) * X  + k_4 * X^2 * Y 
    dY = k_3 * X - k_4 * X^2 * Y
    end k_1 k_2 k_3 k_4 
```

```julia
tspan = (0.0,50.0)
k_1=1.
k_2=1.
k_3=3.
k_4=1.
u₀=[0.0,0.0]
p=[k_1,k_2,k_3,k_4];
```

```julia
prob1 = ODEProblem(brusselator3!,u₀,tspan,p)
sol1 = solve(prob1)
plot(sol1,label=["X","Y"])
title!("Brusselator")
xlabel!("Time [s]")
ylabel!("Concentration [a.u]")
```

A very convenient way to plot these type of oscillatory systems is the phase plot, where we plot one variable versus the other. This is a direct way of looking at the limit cycle. If we plot teh nullclines on top of the limit cycle, we see that the solution cycles around the steady state point (where the nullclines cross).

```julia
p1
plot!(sol1,vars=(1,2),label=["limit cycle plot"])
title!("Brusselator ")
xlabel!("X [a.u.]")
ylabel!("Y [a.u.]")
```
